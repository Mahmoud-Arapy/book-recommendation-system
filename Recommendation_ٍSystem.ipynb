```python
import pandas as pd
import random
from faker import Faker

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

# Initialize Faker and set seeds for reproducibility
1. Generate Books Data (Items Metadata)

```python
fake = Faker('en_US')
Faker.seed(42)
random.seed(42)

NUM_BOOKS = 50
categories = ['Physics', 'Chemistry', 'Computer Science', 'Biology', 'Mathematics']
keywords_map = {
    'Physics': ['quantum', 'mechanics', 'energy', 'relativity', 'thermodynamics'],
    'Chemistry': ['organic', 'inorganic', 'reactions', 'molecules', 'polymers'],
    'Computer Science': ['AI', 'algorithms', 'data structures', 'machine learning', 'networks'],
    'Biology': ['genetics', 'evolution', 'cells', 'anatomy', 'ecology'],
    'Mathematics': ['calculus', 'algebra', 'geometry', 'statistics', 'topology']
}
prefixes = ['Introduction to', 'Advanced', 'Fundamentals of', 'Principles of', 'Applied']

books_data = []
for book_id in range(1, NUM_BOOKS + 1):
    category = random.choice(categories)
    title = f"{random.choice(prefixes)} {category}"
    kws = random.sample(keywords_map[category], random.randint(2, 3))
    
    books_data.append({
        'Book_ID': book_id,
        'Title': f"{title} - Edition {random.randint(1, 4)}",
        'Category': category,
        'Keywords': ", ".join(kws),
        'Publish_Year': fake.year()
    })

df_books = pd.DataFrame(books_data)

```

2. Generate Users Data (User Profiles)

```python
NUM_USERS = 100
users_data = []

for user_id in range(1, NUM_USERS + 1):
    preferred_cats = random.sample(categories, random.randint(1, 3))
    
    users_data.append({
        'User_ID': user_id,
        'Name': fake.name(),
        'Preferred_Categories': ", ".join(preferred_cats)
    })

df_users = pd.DataFrame(users_data)
```

3. Generate Interactions Data (Events)

```python
NUM_INTERACTIONS = 500
actions = ['Viewed', 'Downloaded']
interactions_data = []

for _ in range(NUM_INTERACTIONS):
    user_id = random.randint(1, NUM_USERS)
    book_id = random.randint(1, NUM_BOOKS)
    action = random.choices(actions, weights=[0.3, 0.7])[0] 
    rating = random.choices([1, 2, 3, 4, 5], weights=[0.05, 0.1, 0.2, 0.4, 0.25])[0] if action == 'Downloaded' else None

    interactions_data.append({
        'User_ID': user_id,
        'Book_ID': book_id,
        'Action': action,
        'Rating': rating
    })

df_interactions = pd.DataFrame(interactions_data)
```

### (Content-Based Filtering Model)

```python
df_books['Features'] = df_books['Category'] + " " + df_books['Keywords']
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df_books['Features'])
```

```python
content_similarity = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

Function to get recommendations

```python
def get_content_based_recommendations(book_title, similarity_matrix, df_books, top_n=3):
    try:
        idx = df_books.index[df_books['Title'] == book_title].tolist()[0]
        
        sim_scores = list(enumerate(similarity_matrix[idx]))
        
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        sim_scores = sim_scores[1:top_n+1]
        
        book_indices = [i[0] for i in sim_scores]
        
        return df_books.iloc[book_indices][['Title', 'Category', 'Keywords']]
    except IndexError:
        return "Book not found."
```

```python
sample_book = df_books['Title'].iloc[0] 
print(f"Because you read: '{sample_book}', we recommend:")
print(get_content_based_recommendations(sample_book, content_similarity, df_books))
```

### (Collaborative Filtering Model) 

```python
user_item_matrix = df_interactions.pivot(index='User_ID', columns='Book_ID', values='Rating')
```

```python
user_item_matrix = user_item_matrix.fillna(0)
```

```python
user_similarity = cosine_similarity(user_item_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)
```

 Function to get recommendations based on similar users

```python
def get_collaborative_recommendations(target_user_id, user_item_matrix, user_similarity_df, df_books, top_n=3):

    similar_users = user_similarity_df[target_user_id].sort_values(ascending=False)
    most_similar_user_id = similar_users.index[1]
    similar_user_books = user_item_matrix.loc[most_similar_user_id]
    
    target_user_books = user_item_matrix.loc[target_user_id]
    
    recommendations = []
    for book_id, rating in similar_user_books.items():
        if rating >= 3.0 and target_user_books[book_id] == 0:
            book_info = df_books[df_books['Book_ID'] == book_id]['Title'].values[0]
            recommendations.append(book_info)
            if len(recommendations) == top_n:
                break
                
    if not recommendations:
        return "No collaborative recommendations available (Needs more data/interactions)."
    
    return recommendations
```

Test Collaborative

```python
sample_user = df_interactions['User_ID'].iloc[0]
print(f"Recommendations for User ID {sample_user} based on similar readers:")
recommended_books = get_collaborative_recommendations(sample_user, user_item_matrix, user_similarity_df, df_books)
if isinstance(recommended_books, list):
    for book in recommended_books:
        print(f"- {book}")
else:
    print(recommended_books)
```

### Hybrid Recommendation System (Fallback Strategy)

```python
def get_hybrid_recommendations(target_user_id, top_n=3):
    print(f"\n[ Processing Hybrid Recommendations for User {target_user_id} ]")
    
    collab_result = get_collaborative_recommendations(
        target_user_id, user_item_matrix, user_similarity_df, df_books, top_n
    )
    
    if isinstance(collab_result, list) and len(collab_result) > 0:
        print("Strategy Used: Collaborative Filtering (Found similar readers)")
        return collab_result
        
    print("Collaborative failed (Not enough data). Switching to Content-Based...")
    
    user_history = df_interactions[df_interactions['User_ID'] == target_user_id]
    
    if not user_history.empty:
        best_interaction = user_history.sort_values(by=['Rating', 'Action'], ascending=[False, False]).iloc[0]
        best_book_id = best_interaction['Book_ID']
        best_book_title = df_books[df_books['Book_ID'] == best_book_id]['Title'].values[0]
        
        print(f"🔍 Analyzing content similar to their top book: '{best_book_title}'")
        content_result = get_content_based_recommendations(
            best_book_title, content_similarity, df_books, top_n
        )
        
        if isinstance(content_result, pd.DataFrame):
            return content_result['Title'].tolist()
            
    else:
        print("Absolute Cold Start. User has no history. Using Profile Preferences...")
        
        user_prefs = df_users[df_users['User_ID'] == target_user_id]['Preferred_Categories'].values[0]
        print(f"User prefers: {user_prefs}")
        
        prefs_list = [p.strip() for p in user_prefs.split(',')]
        
        cold_start_recs = df_books[df_books['Category'].isin(prefs_list)].sample(min(top_n, len(df_books)))['Title'].tolist()
        
        return cold_start_recs
```

```python
def get_collaborative_recommendations(target_user_id, user_item_matrix, user_similarity_df, df_books, top_n=3):

    if target_user_id not in user_similarity_df.columns:
        return "No collaborative recommendations available (New user, no interactions yet)."

    similar_users = user_similarity_df[target_user_id].sort_values(ascending=False)
    most_similar_user_id = similar_users.index[1]
    
    similar_user_books = user_item_matrix.loc[most_similar_user_id]
    target_user_books = user_item_matrix.loc[target_user_id]
    
    recommendations = []
    for book_id, rating in similar_user_books.items():
        if rating >= 3.0 and target_user_books[book_id] == 0:
            book_info = df_books[df_books['Book_ID'] == book_id]['Title'].values[0]
            recommendations.append(book_info)
            if len(recommendations) == top_n:
                break
                
    if not recommendations:
        return "No collaborative recommendations available (Needs more data/interactions)."
    
    return recommendations
```

Testing the Hybrid System

```python
test_user = 33 
hybrid_books_existing = get_hybrid_recommendations(test_user)

print(f"\n🎯 Final Recommendations for User {test_user}:")
if isinstance(hybrid_books_existing, list):
    for i, book in enumerate(hybrid_books_existing, 1):
        print(f"{i}. {book}")
else:
    print(hybrid_books_existing)

print("\n" + "="*50)

print("\n--- 2. Testing Absolute Cold Start (Completely New User) ---")
new_user_id = 999
if new_user_id not in df_users['User_ID'].values:
    df_users.loc[len(df_users)] = [new_user_id, 'New Student', 'Physics, Biology']

new_user_books = get_hybrid_recommendations(new_user_id)

print(f"\n🎯 Final Recommendations for New User {new_user_id}:")
if isinstance(new_user_books, list):
    for i, book in enumerate(new_user_books, 1):
        print(f"{i}. {book}")
else:
    print(new_user_books)
```
